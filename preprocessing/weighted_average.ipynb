```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/BA_data/찬우/solar_split_raw/yeongheung.csv')
data
```

```python
data['발전기'].value_counts()
```

```python
data.columns
```

## Drop unnecessary columns

```python
data = data.drop('Unnamed: 0.1', axis=1)
data = data.drop('Unnamed: 0', axis=1)
data = data.drop('총량(MW)', axis=1)
data = data.drop('평균(MW)', axis=1)
data = data.drop('최대(시간별)', axis=1)
data = data.drop('최소(시간별)', axis=1)
data = data.drop('최대', axis=1)
data = data.drop('최소', axis=1)
data = data.drop('sunrise', axis=1)
data = data.drop('sunset', axis=1)
data = data.drop('region', axis=1)
data
```

```python
data.groupby('date_only')['발전기'].unique()
```

## Create 'capacity' column

```python
import numpy as np
# Generate 'capacity' column
data['capacity'] = np.where(data['발전기'] == '영흥태양광 1', 978,
                          np.where(data['발전기'] == '영흥태양광 #3 1', 978,
                                   np.where(data['발전기'] == '영흥태양광 #3 2', 978,
                                            np.where(data['발전기'] == '영흥태양광 #3 3', 1028,
                                                     np.where(data['발전기'] == '영흥태양광 2', 978,
                                                              np.where(data['발전기'] == '영흥태양광#5 1', 978, 0))))))
data
```

```python
# Check NaN value in 'capacity' column
data['capacity'].isna().any()
```

## Create 'weighted_average_power' column

```python
# Generate 'weighted_average_power' column
data['weighted_average_power'] = (data['발전량(MWh)'] * data['capacity']) / data.groupby(['date_only', '발전기'])['capacity'].transform('sum')
data
```

## Create 'total_weighted_average_power' column

```python
# Add all 'weighted_average_power' of all units for the same time
data['total_weighted_average_power'] = data.groupby('date')['weighted_average_power'].transform('sum')

# Output result
data[['date', '발전기', 'weighted_average_power', 'total_weighted_average_power']]
```

```python
data[data['date']=='2022-01-31 08:00:00']
```

```python
data.columns
```

## Create new dataframe

```python
# Create new dataframe
new_df = data.drop_duplicates('date')[['date', 'date_only', 'temperature', 'precipitation', 'wind speed', 'humidity', 'air pressure',
                                     'sunshine', 'insolation', 'snowfall', 'cloud', 'ground temperature', 'total_weighted_average_power']]

new_df
```

## Specify 'yeongheung'

```python
# Add 'power_plant' column
new_df['power_plant'] = '영흥발전소'
new_df
```

## Save Dataframe

```python
# Save new_df as csv foramt file
new_df.to_csv('/content/drive/MyDrive/BA_data/찬우/solar_integrated_raw/yeongheung_di.csv', index=False)
```
