```python
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/하늘/data files/merged_dataset.csv')
```

```python
generator_set = ["samcheonpo","yeongheung","yeongdong","gumi","gwangyanghang","dusan","gyeongsangUni","yecheon","goheungman"]
```

```python
# disunite dataframe into 9 different dataframes

samcheonpo_df = data[data['발전기'].str.contains('삼천포')]
yeongheung_df = data[data['발전기'].str.contains('영흥')]
yeongdong_df = data[data['발전기'].str.contains('영동')]
gumi_df = data[data['발전기'].str.contains('구미')]
gwangyanghang_df = data[data['발전기'].str.contains('광양항')]
dusan_df = data[data['발전기'].str.contains('두산')]
gyeongsangUni_df = data[data['발전기'].str.contains('경상대')]
yecheon_df = data[data['발전기'].str.contains('예천')]
goheungman_df = data[data['발전기'].str.contains('고흥만')]

df_set = [samcheonpo_df, yeongheung_df, yeongdong_df, gumi_df, gwangyanghang_df, dusan_df, gyeongsangUni_df, yecheon_df, goheungman_df]
```

```python
# export dataframe to csv file

for i in range(9):
  df = df_set[i]
  file_name = generator_set[i]+".csv"
  df.to_csv(file_name)
```
