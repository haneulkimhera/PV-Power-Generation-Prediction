```python
days_in_month = {1: 31, 2: 28, 3: 31, 4: 30, 5: 31, 6: 30, 7: 31, 8: 31, 9: 30, 10: 31, 11: 30, 12: 31}
```

```python
# Feature Generation - month_x, month_y, day_x, day_y, hour_x, hour_y

for generator in generator_set:

  path = '/content/drive/MyDrive/하늘/data files/nullcheck_interpolated/'+generator+'.csv'
  file_path = '/content/drive/MyDrive/하늘/data files/feature_added/'+generator+'.csv'
  data = pd.read_csv(path)

  data = data.drop('Unnamed: 0', axis=1)
  data = data.drop('date_only', axis=1)
  print(data.shape)

  data['year'] = pd.to_datetime(data['date']).dt.year
  data['month'] = pd.to_datetime(data['date']).dt.month
  data['day'] = pd.to_datetime(data['date']).dt.day
  data['hour'] = pd.to_datetime(data['date']).dt.hour

  data['DoM'] = data['month'].map(days_in_month)

  # make a new feature 'month_x' = sin(month * (2pi/12))
  data['month_x'] = np.sin(data['month'] * (2 * np.pi / 12))
  # make a new feature 'month_y' = cos(month * (2pi/12))
  data['month_y'] = np.cos(data['month'] * (2 * np.pi / 12))
  # make a new feature 'day_x' = in(day * (2pi/DoM))
  data['day_x'] = np.sin(data['day'] * (2 * np.pi / data['DoM']))
  # make a new feature 'day_y' = cos(day * (2pi/DoM))
  data['day_y'] = np.cos(data['day'] * (2 * np.pi / data['DoM']))
  # make a new feature 'hour_x' = sin(hour * (2pi/24))
  data['hour_x'] = np.sin(data['hour'] * (2 * np.pi / 24))
  # make a new feature 'hour_y' = cos(hour * (2pi/24))
  data['hour_y'] = np.cos(data['hour'] * (2 * np.pi / 24))

  data.to_csv(file_path)
```
